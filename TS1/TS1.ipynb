### Actividad
Se pedia realizar un generador de señales senoidales parametrizables.

```python
import numpy as np
import matplotlib.pyplot as plt


#N = 1000
#fs = 1000

def my_senoidal (N, freq_M, amplitud = 1, valor_medio = 0, freq = 1, fase = 0):
    
    ts = 1/freq_M
    
    tt = np.linspace(0, (N-1)*ts, N)
    
    xx = amplitud * np.sin(2*np.pi*(freq)*tt + fase)
    
    return(tt,xx)

[t, y] = my_senoidal(N=100,freq_M = 100)

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)

ax.plot(t,y)
```

La señal senoidal se encuentra normalizada. Podemos ver que la forma de salida depende ampliamente de los parametros que definamos.
